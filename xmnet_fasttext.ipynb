```python
from mxnet.contrib import text
```

```python
text.embedding.get_pretrained_file_names()
```

```python
text_data = " 你好世界 \n 你好善良的世界 \n 嘿世界 \n 你好世界"
counter = text.utils.count_tokens_from_str(text_data)
```

```python
counter
```

```python
my_vocab = text.vocab.Vocabulary(counter)
```

```python
my_embedding = text.embedding.create('fasttext', pretrained_file_name='wiki.zh.vec',vocabulary=my_vocab)
```

```python
 my_embedding.get_vecs_by_tokens('你好世界')
```
