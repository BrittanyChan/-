```python
import pandas as pd

df = pd.read_csv(r'C:\Users\yanyuyao\Desktop\chenxi\kmeans_test.csv',encoding='gbk')
```

```python
df.head()
```

```python
df1 = df.loc[:,['Max-min Scalar1','Max-min Scalar2','Max-min Scalar3','Max-min Scalar4']]
df1.head()
```

```python
df_values = df1.values
```

```python
import matplotlib.pyplot as plt  
from sklearn.cluster import KMeans

estimator = KMeans(n_clusters=3)#构造聚类器
estimator.fit(df_values)#聚类
label_pred = estimator.labels_ #获取聚类标签
```

```python
x0 = df_values[label_pred == 0]
x1 = df_values[label_pred == 1]
x2 = df_values[label_pred == 2]
```

```python
import matplotlib.pylab as pylab

params={
    'axes.labelsize': '35',       
    'xtick.labelsize':'27',
    'ytick.labelsize':'27',
    'lines.linewidth':2 ,
    'legend.fontsize': '27',
    'figure.figsize'   : '40,20'
}
pylab.rcParams.update(params)            #set figure parameter

plt.subplot(231)
plt.scatter(x0[:, 0], x0[:, 1], c = "red", marker='o', label='label0', linewidths = 10)
plt.scatter(x1[:, 0], x1[:, 1], c = "green", marker='*', label='label1', linewidths = 10)  
plt.scatter(x2[:, 0], x2[:, 1], c = "blue", marker='^',label='label2', linewidths = 10)  
plt.xlabel('Max-min1')  
plt.ylabel('Max-min2-3')  
plt.legend(loc=2)

plt.subplot(232)
plt.scatter(x0[:, 0], x0[:, 2], c = "red", marker='o', label='label0', linewidths = 10)  
plt.scatter(x1[:, 0], x1[:, 2], c = "green", marker='*', label='label1', linewidths = 10)  
plt.scatter(x2[:, 0], x2[:, 2], c = "blue", marker='^', label='label2', linewidths = 10)  
plt.xlabel('Max-min1')  
plt.ylabel('Max-min4-7')  
plt.legend(loc=2) 

plt.subplot(233)
plt.scatter(x0[:, 0], x0[:, 3], c = "red", marker='o', label='label0', linewidths = 10)  
plt.scatter(x1[:, 0], x1[:, 3], c = "green", marker='*', label='label1', linewidths = 10)  
plt.scatter(x2[:, 0], x2[:, 3], c = "blue", marker='^', label='label2', linewidths = 10)  
plt.xlabel('Max-min1')  
plt.ylabel('Max-min8')  
plt.legend(loc=2) 

plt.subplot(234)
plt.scatter(x0[:, 1], x0[:, 2], c = "red", marker='o', label='label0', linewidths = 10)  
plt.scatter(x1[:, 1], x1[:, 2], c = "green", marker='*', label='label1', linewidths = 10)  
plt.scatter(x2[:, 1], x2[:, 2], c = "blue", marker='^', label='label2', linewidths = 10)  
plt.xlabel('Max-min2-3')  
plt.ylabel('Max-min4-7')  
plt.legend(loc=2) 

plt.subplot(235)
plt.scatter(x0[:, 1], x0[:, 3], c = "red", marker='o', label='label0', linewidths = 10)  
plt.scatter(x1[:, 1], x1[:, 3], c = "green", marker='*', label='label1', linewidths = 10)  
plt.scatter(x2[:, 1], x2[:, 3], c = "blue", marker='^', label='label2', linewidths = 10)  
plt.xlabel('Max-min2-3')  
plt.ylabel('Max-min8')  
plt.legend(loc=2)

plt.subplot(236)
plt.scatter(x0[:, 2], x0[:, 3], c = "red", marker='o', label='label0', linewidths = 10)  
plt.scatter(x1[:, 2], x1[:, 3], c = "green", marker='*', label='label1', linewidths = 10)  
plt.scatter(x2[:, 2], x2[:, 3], c = "blue", marker='^', label='label2', linewidths = 10)  
plt.xlabel('Max-min4-7')  
plt.ylabel('Max-min8')  
plt.legend(loc=2)

plt.savefig('kmeans_test.png',dpi=200)
plt.show()
```
